สำหรับครั้งนี้สิ่งที่เพิ่มเข้ามาแล้วสำคัญมากกกกคือคำสั่ง `import` <br>
จากตัวอย่างจะเห็นอยู่ด้านบนสุด คือ `import random`  #เพราะ python รันคำสั่งจากบนลงล่าง <br> 
ซึ่งในส่วนนี้ขออธิบายคำศัพท์ที่สำคัญเพิ่มคำนึงคือ `library` หรือที่แปลว่าห้องสมุดแหละ <br>

---

โดยปกติแล้วลักษณะการเขียน (syntax) ของคำสั่ง import คือ `import <library>` <br>
library คือชุดคำสั่งที่เราดึงนำมาใช้ ให้นึกภาพว่ามีโปรแกรมเมอร์เทพๆเขียนคำสั่งให้เรา แล้วเขาก็ตั้งชื่อ(library)ไว้แล้วเปิดฟรีให้คนมาใช้งานได้ <br>
ซึ่งในฐานะมนุษย์ทั่วไปๆอย่างเราๆก็อยากได้โค้ดนั้นๆมาประยุกต์กับงานของเรา ดังนั้นนี่แหละสาเหตุว่าทำไมต้อง `import`

ในตัวอย่างจะใช้เป็น library ที่มีชื่อว่า `random` ซึ่งมันทำให้เราสามารถใช้คำสั่งในการสุ่มได้ โดยในที่นี้เราประยุกต์การสุ่มกับการจะให้คู่ต่อสู้สุ่มว่าตอบเป็นค้อน กระดาษ หรือกรรไกร <br>
จริงๆแล้ว library random ยังทำอื่นๆได้อีกมากมายสนใจลอง search google ว่า `python random library documentation`

```python
import random # library ชื่อ random 
 
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''


choice = (input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors\n0/1/2:")) # ผู้ใช้งานกรอกว่าจะออกค้อน กระดาษ กรรไกร
computers_choice = random.randint(0,2) # การสุ่มจาก random library โดยคำสั่งดังกล่าวจำทำหน้าที่สุ่มเลขออกมา ในที่นี้กำหนดว่าสุ่มเลข [0,1,2]

if not choice.isdigit():# เพื่อเป็นการตรวจว่า'ทุกตัวอักษรที่กรอกมา'เป็นตัวเลขหรือไม่ *ตัวอย่างเพิ่มเติมด้านล่าง
  print("You've entered an invalid value, try again and choose a number between 0-2.")
else:
  choice = int(choice) # แปลงคำตอบจาก data type ประเภท ตัวอักษร(str)เป็นตัวเลขจำนวนเต็ม(int) *ซึ่งในที่นี้จะแปลงได้แน่นอน 100% เพราะจากคำสั่ง .isdigit() จะเป็นการเช็คแล้วว่าเป็นตัวเลขทุกตัวไหม
  if choice > 2:  # ถ้าหากตัวแปร choice มากกว่า 2 แปลว่าคนกรอกมันกรอกไม่ถูกละ เพราะเราตั้งใจให้กรอก 0, 1, 2 ดังนั้นถ้ามากกว่า 2 เราจะให้เขาไปกรอกใหม่
    print("You've entered an invalid number, try again and choose a number between 0-2.")

    ### code หลังจากนี้เป็นแค่กฎของเกมค้อน กระดาษ กรรไกรปกติ ###
  elif choice == 0 or choice == 1 or choice == 2:
    
    if choice == 0: # หากเราเลือก 0 = ค้อน 
      print(f"You chose: Rock {rock}")
    elif choice == 1: # หากเราเลือก 1 = กระดาษ 
      print(f"You chose: Paper {paper}")
    elif choice == 2: # หากเราเลือก 2 = กรรไกร 
      print(f"You chose: Scissors {scissors}")
   
   
    if computers_choice == 0: # หากคู่ต่อสู้เลือก 0 = ค้อน 
      print(f"The computer chose: Rock {rock}") 
      if choice == 2: # ถ้าเราเลือก 2 = กรรไกร ก็จะแปลว่าเราแพ้ เพราะค้อน(ฝ่ายตรงข้าม)ชนะกรรไกร(ของเรา)
        print("You lose, Rock wins against scissors.")
      elif choice == 0: # ถ้าเราเลือก 0 = ค้อน ก็จะเสมอ
        print("It's a draw!")
      else: # ถ้าไม่เข้าเงื่อนไขอะไรเลยแปลว่าเราชนะ (จริงๆถ้าเขียน elif ก็จะเขียนว่า elif code == 1 หละ)
        print("You win!")

    elif computers_choice == 1:
      print(f"The computer chose: Paper {paper}")
      if choice == 0:
        print("You lose, Paper wins against rock.")
      elif choice == 0:
        print("It's a draw!")
      else:
        print("You win!")

    elif computers_choice == 2:
      print(f"The computer chose: Scissors {scissors}")
      if choice == 1:
        print("You lose, Scissors win against paper.")
      elif choice == 2:
        print("It's a draw!")
      else:
        print("You win!")
```

```python
## อ่านเพิ่มเติมเรื่อง .isdigit() ->  https://www.w3schools.com/python/ref_string_isdigit.asp

# จะเห็นว่าทุกตัวอักษรเป็นตัวเลขจึงตอบ True
txt = "50800"
x = txt.isdigit()
print(x)

# มีตัวอักษร x หนึ่งตัวที่ไม่ใช้ตัวเลขจึงตอบ False 
txt = "50x800"
x = txt.isdigit()
print(x)

# มีตัวอักษร - หนึ่งตัวที่ไม่ใช้ตัวเลขจึงตอบ False 
txt = "-50800"
x = txt.isdigit()
print(x)
```
