```python
# เรียงลำดับผิด โดยอันนี้เรียงเป็น if -> else -> elif 
text = 'test'
if text == 'text':
    print('if')
else:
    print('else')
elif text == 'test':
    print('elif')
```

```python
# เรียงลำดับผิด โดยอันนี้เรียงเป็น  else -> if -> elif 
text = 'test'
else:
    print('else')
if text == 'text':
    print('if')
elif text == 'test':
    print('elif')
```

```python
# เรียงลำดับถูก (จะเห็นว่าขอแค่เรียงถูกไม่ต้องมี else ก็ได้)
text = 'test'
if text == 'text':
    print('if')
elif text == 'test':
    print('elif')
```

```python
# เรียงลำดับถูก (จะเห็นว่ารอบนี้เข้า elif อันที่ 2 เพราะ text = 'test2' ซึ้งตรงกับเงื่อนไขอันที่ 2)
text = 'test2'
if text == 'text':
    print('if')
elif text == 'test':
    print('elif')
elif text == 'test2':
    print('elif2')
```

```python
# ครั้งนี้จะเห็นว่า else ไม่ต้องเขียนเงื่อนไขอะไรเหมือน if, elif ซึ่งมันจะทำงานเสมอถ้าหากไม่เข้าเงื่อนไขใดๆที่กำหนดก่อนหน้าเลย
text = 'welcome'
if text == 'text':
    print('if')
elif text == 'test':
    print('elif')
elif text == 'test2':
    print('elif2')
else:
    print('else')
```
